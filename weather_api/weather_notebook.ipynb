```python
import io
import pandas as pd
import matplotlib.pyplot as plt
from app.api_client import get_station_metadata, get_station_measurements

""" metadata_csv_path = get_station_metadata("SE01")
df = pd.DataFrame([metadata_csv_path])
df.head() """

data = get_station_measurements("SE01", 2024, 1, 1)
df = pd.read_csv(io.StringIO(data), sep='\t', header=None, 
                names=['time', 'temperature', 'precipitation'])
df.head()
```

```python
plt.figure(figsize=(20, 10))
plt.plot(df['time'], df['temperature'], '-o', markersize=3)
plt.title('Temperature on January 1, 2024 - Station SE01')
plt.xlabel('Time')
plt.ylabel('Temperature (C)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
